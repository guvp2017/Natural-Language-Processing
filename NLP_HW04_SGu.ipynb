```python
## import libraries

import pandas as pd
import numpy as np

import nltk
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag
from nltk.corpus import brown
from nltk import UnigramTagger
from nltk import StanfordPOSTagger

import os
wd =os.getcwd()

import spacy
import en_core_web_sm

import pattern
from pattern.en import tag
```

```python
## make pos_spacy() to call spaCy tagger
def pos_spacy(sentence):
    nlp = en_core_web_sm.load()
    doc = nlp(sentence)
    tagged = [(token.text, token.tag_) for token in doc]    
    return tagged
```

```python
## make pos_stanford() to call Stanford POS tagger
def pos_stanford(tokens):     
    model = wd+'/stanford-postagger-2018-10-16/models/english-bidirectional-distsim.tagger'
    jar = wd+'/stanford-postagger-2018-10-16/stanford-postagger.jar'
    Stanfordpos_tagger = StanfordPOSTagger(model, jar, encoding='utf-8')    
    stanford = Stanfordpos_tagger.tag(tokens)    
    return(stanford)
```

```python
## make pos_uniform() to call Unigram POS tagger
def pos_unigram(tokens):     
    brown_tagged_sents = brown.tagged_sents(categories='news')      
    unigram_tagger = UnigramTagger(brown_tagged_sents) # train unigram tagger     
    unigram = unigram_tagger.tag(tokens)  
    return(unigram)
```

```python
## make pos_tagger() to call NLTK, spacy, pattern, stanford, or unigram POS tagger

def pos_tagger (sentence, tagger):     
    
    penn_pos = pd.read_csv("penn_treebank.csv") # Use Penn Treebank tagset
    
    if tagger == 'nltk': # NLTK POS tagger
        tokens = nltk.word_tokenize(sentence)
        tag_nltk = nltk.pos_tag(tokens)
        
        df_tag = pd.DataFrame(tag_nltk, columns = ['word', 'pos'])
        df_desc = pd.merge(df_tag, penn_pos, on = 'pos', how = 'left')
        df_desc.columns = ['word', 'pos', 'desc']    
        
    elif tagger == 'spacy': # spaCy POS tagger             
        tag_spacy = pos_spacy(sentence)
        
        df_tag = pd.DataFrame(tag_spacy, columns = ['word', 'pos'])
        df_desc = pd.merge(df_tag, penn_pos, on = 'pos', how = 'left')
        df_desc.columns = ['word', 'pos', 'desc'] 
        
    elif tagger == 'pattern': # Pattern POS tagger
        tag_ptrn = tag(sentence)
        
        df_tag = pd.DataFrame(tag_ptrn, columns = ['word', 'pos'])
        df_desc = pd.merge(df_tag, penn_pos, on = 'pos', how = 'left')
        df_desc.columns = ['word', 'pos', 'desc'] 
        
    elif tagger == 'stanford': # Stanford POS tagger
        tokens = nltk.word_tokenize(sentence)
        tag_nltk = nltk.pos_tag(tokens)
        tag_stanford = pos_stanford(tokens)
        
        df_tag = pd.DataFrame(tag_stanford, columns = ['word', 'pos'])
        df_desc = pd.merge(df_tag, penn_pos, on = 'pos', how = 'left')
        df_desc.columns = ['word', 'pos', 'desc']
    
    elif tagger == 'unigram': # Unigram POS tagger
        tokens = nltk.word_tokenize(sentence)
        tag_unigram = pos_unigram(tokens)      
        
        df_tag = pd.DataFrame(tag_unigram, columns = ['word', 'pos'])
        df_desc = pd.merge(df_tag, penn_pos, on = 'pos', how = 'left')
        df_desc.columns = ['word', 'pos', 'desc']       
                
    return (df_desc)
```
